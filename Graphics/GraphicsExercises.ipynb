```python
# Plot of the Lorenz Attractor based on Edward Lorenz's 1963 "Deterministic
# Nonperiodic Flow" publication.
# http://journals.ametsoc.org/doi/abs/10.1175/1520-0469%281963%29020%3C0130%3ADNF%3E2.0.CO%3B2
#
# Note: Because this is a simple non-linear ODE, it would be more easily
#       done using SciPy's ode solver, but this approach depends only
#       upon NumPy.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def lorenz(x, y, z, s=7, r=30, b=4) :
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01
stepCnt = 8000

# Need one more for the initial values
xs = np.empty((stepCnt + 1,))
ys = np.empty((stepCnt + 1,))
zs = np.empty((stepCnt + 1,))

# Setting initial values
xs[0], ys[0], zs[0] = (0.5, 1.1, 1.07)

# Stepping through "time".
for i in range(stepCnt) :
    # Derivatives of the X, Y, Z state
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(xs, ys, zs, color="green", linestyle='-.')
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor", color="blue")
ax.set_xlim([-20,20])
ax.set_ylim([-20,20])

plt.show()

```

**5.** Determine the electric potential for a quadrupole consisting of

$q1$ = -1.0 nC at (-2 m, 0 m)

$q2$ = -1.0 nC at (+2 m, 0 m)

$q3$ = +1.0 nC at (0 m, -2 m)

$q4$ = +1.0 nC at (0 m, +2 m)

**(a)** Make a surface plot on the range $-4 <(x,y)< 4$ using 101 bins in $x$ and 101 bins in $y$. In this case, do not let $k=q=d=1$. Use dimensionful quantities and compute the value of the potential in volts.
Recall $k = 8.9875517873681764 \times 10^9 N\cdot\frac{m^2}{C^2}$.

```python
from mpl_toolkits.mplot3d import Axes3D

dx = 0.06
dy = 0.06
x = np.arange(-3, 3, dx)
y = np.arange(-3, 3, dy)
x, y = np.meshgrid(x, y)

k = 8.9875517873681764e9; q = 1e-9; d = 2
V=(k*q/(x**2+(y-d)**2)**(1/2.)) + (k*q/(x**2+(y+d)**2)**(1/2.)) - (k*q/((x+d)**2+y**2)**(1/2.)) - (k*q/((x-d)**2+y**2)**(1/2.))

fig = plt.figure(figsize=(16,9))

ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, V, rstride=1, cstride=1, linewidth=0,cmap=cm.spectral)
cb = fig.colorbar(p, shrink=0.5)

title("Electric Potential of Four Charges", fontsize=20)
xlabel("x", fontsize=16)
ylabel("y", fontsize=16)
ax.set_zlabel("V", fontsize=16)

plt.show()
```

**(b)** Now make a contour plot of the same data, with appropriate contour levels to see the shape of the potential.

```python
levels = array([-45,-20,-12,-8,-4,-2,-1,-.5,0,.5,1,2,4,8,12,20,45])
plt.figure(figsize=(8,8))
plt.title('Electric Potential Contour Plot', fontsize=20)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.contour(x,y,V,levels,linewidths=4,cmap=cm.coolwarm);
```

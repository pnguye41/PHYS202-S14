```python
# Plot of the Lorenz Attractor based on Edward Lorenz's 1963 "Deterministic
# Nonperiodic Flow" publication.
# http://journals.ametsoc.org/doi/abs/10.1175/1520-0469%281963%29020%3C0130%3ADNF%3E2.0.CO%3B2
#
# Note: Because this is a simple non-linear ODE, it would be more easily
#       done using SciPy's ode solver, but this approach depends only
#       upon NumPy.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def lorenz(x, y, z, s=7, r=30, b=4) :
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot


dt = 0.01
stepCnt = 8000

# Need one more for the initial values
xs = np.empty((stepCnt + 1,))
ys = np.empty((stepCnt + 1,))
zs = np.empty((stepCnt + 1,))

# Setting initial values
xs[0], ys[0], zs[0] = (0.5, 1.1, 1.07)

# Stepping through "time".
for i in range(stepCnt) :
    # Derivatives of the X, Y, Z state
    x_dot, y_dot, z_dot = lorenz(xs[i], ys[i], zs[i])
    xs[i + 1] = xs[i] + (x_dot * dt)
    ys[i + 1] = ys[i] + (y_dot * dt)
    zs[i + 1] = zs[i] + (z_dot * dt)

fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(xs, ys, zs, color="green", linestyle='-.')
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("Lorenz Attractor", color="blue")
ax.set_xlim([-20,20])
ax.set_ylim([-20,20])

plt.show()

```

```python
x = np.arange(3142)

sunspots = loadtxt('sunspots.txt', unpack=True, usecols=[1])
y = sunspots[x]

plt.figure(figsize=(16, 9), dpi=80)
plt.plot(x, y, '-', markersize = 2, color='orangered', label='Sunspots')
plt.xticks([12, 312, 612, 912, 1212, 1512, 1812, 2112, 2412, 2712, 3012],[r'$1750$', r'$1775$', r'$1800$', r'$1825$', r'$1850$', r'$1875$', r'$1900$', r'$1925$', r'$1950$', r'$1975$', r'$2000$'])
plt.yticks([0, 50, 100, 150, 200, 250, 300])
plt.xlim([-30,3172])
plt.ylim([0,290])
plt.xlabel("Year", color='b', fontsize=16)
plt.ylabel("Sunspots", color='r', fontsize=16)
plt.title("Monthly Sunspots Observed between Jan 1749 & Oct 2010", fontsize=20)


peak = sunspots.max()
peak_date_float = (np.where(sunspots==peak)[0][0]) / 12.
peak_year = 1749 + int(peak_date_float)
peak_month = int((peak_date_float - int(peak_date_float))*12)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
peak_date = months[peak_month] + ' ' + str(peak_year)

plt.scatter([peak_date_float*12., ], [peak, ],25,  color='black', label='Peak')
plt.annotate(r'Highest peak: %s sunspots in %s' % (peak, peak_date),
             xy=[peak_date_float*12, peak],
             xytext=[peak_date_float*12 - 450, peak + 20], fontsize=12,
             arrowprops=dict(arrowstyle="->", color='b'))

plt.legend(loc='upper center')

plt.grid()
```

**(b)** Now add code to calculate the running average of the data. Plot both the original data and the running average on the same graph, only over the range covered by the first 1000 data points.

```python
x = np.arange(1000)

sunspots = loadtxt('sunspots.txt', unpack=True, usecols=[1])
y = sunspots[x]

ravg = []
for n in x+1:
    ravg.append(int(np.average(sunspots[:n])))
y2 = ravg

plt.figure(figsize=(16, 9), dpi=80)
plt.plot(x, y, '-', markersize = 2, color='orangered', label='Sunspots')
plt.plot(x, y2, '-', markersize = 4, color='g', label='Running Average')
plt.xticks([12, 312, 612, 912, 1212, 1512, 1812, 2112, 2412, 2712, 3012],[r'$1750$', r'$1775$', r'$1800$', r'$1825$', r'$1850$', r'$1875$', r'$1900$', r'$1925$', r'$1950$', r'$1975$', r'$2000$'])
plt.yticks([0, 50, 100, 150, 200, 250, 300])
plt.xlim([-30,1000])
plt.ylim([0,290])
plt.xlabel("Year", color='b', fontsize=16)
plt.ylabel("Sunspots", color='r', fontsize=16)
plt.title("Monthly Sunspots Observed between Jan 1749 & Oct 2010", fontsize=20)


peak = sunspots[:1000].max()
peak_date_float = (np.where(sunspots[:1000]==peak)[0][0]) / 12.
peak_year = 1749 + int(peak_date_float)
peak_month = int((peak_date_float - int(peak_date_float))*12)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
peak_date = months[peak_month] + ' ' + str(peak_year)

plt.scatter([peak_date_float*12., ], [peak, ],25,  color='black', label='Peak')
plt.annotate(r'Highest peak: %s sunspots in %s' % (peak, peak_date),
             xy=[peak_date_float*12, peak],
             xytext=[peak_date_float*12 - 300, peak + 20], fontsize=12,
             arrowprops=dict(arrowstyle="->", color='b'))

plt.legend(loc='upper center')

plt.grid()
```

## Monte Carlo Integration

MC integration uses a random sample of points within a domain where our funciton is defined to estimate the bounded area. It is actually easier to explain with an analogy in 2 dimensions. Imagine a circular dartboard which I want to know the area of, hanging on a rectangular board. If I throw $N$ darts randomly at it, the area of the circle is approximate the ratio of darts inside it, $N_{in}$, to the total darts thrown, $N=N_{in}+N_{out}$ times the area of the rectangular board:
\begin{equation}
A_{circ}\simeq \frac{N_{in}}{N} A_{board}
\\
A_{board} = l \times w
\\
A_{exact} = \pi r^2
\end{equation}

```python
l = w = 4.
r = 1.375
Nin = 15
N = 40
Aboard = l*w
Aexact = pi*(r**2)
Acirc = (Nin/float(N))*Aboard #since we calc a ratio, make sure y's are floats!
print "Aboard =",Aboard
print "Aexact =",Aexact
print "Acirc  =",Acirc
print "Error  =",(100.*(Acirc-Aexact)/Aexact),"%"
```

For better approximation, increase the number of darts. To improve speed/efficiency, we can shrink the size of the board so it is just big enough for the dartboard. This is known as the "Hit of Miss" MC method. Let's outline how to implement it:

### Hit of Miss MC Integration

1. Define function to integrate (or array of points)
2. Create bounding box around function within which to work
3. Randomly (or selectively randomly) sample points in region
4. Calc fraction of points that fall in function boundary
5. Integral is point fraction times area of bounding box

We'll work with: $f(x) = x^{1/2} + \sin(x)$

**Step1:**

```python
f = lambda x: x**(0.5) + np.sin(x)
```

```python
x = np.arange(0.,15.,0.001)
```

```python
plt.plot(x,f(x),lw=3)
plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
plt.xlim(0.,15.)

plt.show()
```

**Step 2:** Create the bounding box for random sampling. Let's use the limits $(1.7, 12.4)$.

```python
xmin = 1.7
xmax = 12.4

#find max value of f within boundary to set ymax
subx = x[logical_and((x > xmin),(x < xmax))]
ymax = f(subx).max()*1.05
ymin = 0.

#Note use of logical_and function to be able to set mask with two constraints
```

```python
#Replot function
plt.plot(x,f(x),lw=3)
plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
plt.xlim(0.,15.)

#plot box
plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
#vertical left
plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
#vertical right
plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
#horizontal top

plt.title(r'$MC \ Integration$', fontsize=18)
plt.show()
```

**Step 3:** Create random sample points using `random_sample()` function. It returns numbers within [0.,1.).

To get values within our limits, we need to scale and translate the values into our range, as suggested in `help(random_sample)`.

```python
N = 10000
samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
samples_y = ymin + (ymax-ymin)*np.random.random_sample(N);
```

```python
#Replot function
plt.plot(x,f(x),lw=3)
plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
plt.xlim(0.,15.)

#Plotting all of them is not recommended
#plt.plot(samples_x,samples_y,'g.')

#Just show the first 500 points or it gets hard to seee
plt.plot(samples_x[:500],samples_y[:500],'g.')

#plot box
plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
#vertical left
plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
#vertical right
plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
#horizontal top

plt.title(r'$MC \ Integration$', fontsize=18)
plt.show()
```

**Step 4:** Now we need to keep track of the points where y < f(x) and replot them to be sure.

```python
newmask = (samples_y < f(samples_x))
newmask.sum()
```

```python
#Replot function
plt.plot(x,f(x),lw=3)
plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
plt.xlabel(r'$x$',fontsize=20)
plt.xlim(0.,15.)

#Plotting all of them is not recommended
#plt.plot(samples_x,samples_y,'g.')

#Just show the first 500 points or it gets hard to seee
plt.plot(samples_x[:500],samples_y[:500],'g.')

#Use the mask to show the ones below the function in red
plt.plot(samples_x[newmask[:500]],samples_y[newmask[:500]],'r.')

#plot box
plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
#vertical left
plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
#vertical right
plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
#horizontal top

plt.title(r'$MC \ Integration$', fontsize=18)
plt.show()
```

**Step 5:** Now compute area of the rectangle times the ratio of red and green dots.

```python
I = (xmax-xmin)*(ymax-ymin)*newmask.sum()/newmask.size
print I
```

### Comparing to Exact Solution

\begin{equation}
\int_{x_{min}}^{x_{max}} f(x)dx = \int_{x_{min}}^{x_{max}} x^{1/2} dx + \int_{x_{min}}^{x_{max}} \sin(x) dx = \left[\frac{2}{3} x^{3/2} - \cos(x)\right]_{x_{min}}^{x_{max}}
\end{equation}

```python
Iana = lambda x: (2./3.)*x**(3./2.) - np.cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = 100*np.abs(Iexact - I) / Iexact

print "I_exact =",Iexact
print "I_hitormiss =",I
print "Error on I =",Ierr,"%"
```

How many points would it take to reduce the error below 0.001%?
Let's put out work into an interact object and explore the result as a function of $N$.

```python
from IPython.html.widgets import interact, interactive
```

```python
def hitOrMiss(N=10000,xmin=1.7,xmax=12.4):
    
    #Set up function
    f = lambda x: x**(0.5) + np.sin(x)
    x = np.arange(0.,15.,0.001)
    Iana = lambda x: (2./3.)*x**(3./2.) - np.cos(x)
    Iexact = Iana(xmax) - Iana(xmin)
    
    #find max value of f within boundary to set ymax
    subx = x[logical_and((x > xmin),(x < xmax))]
    ymax = f(subx).max()*1.05
    ymin = 0.
    
    #######################################
    #Integration Part
    samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
    samples_y = ymin + (ymax-ymin)*np.random.random_sample(N);
    
    newmask = (samples_y < f(samples_x))
    I = (xmax-xmin)*(ymax-ymin)*newmask.sum()/newmask.size
    Ierr = 100*np.abs(Iexact - I) / Iexact
    ########################################
    
    #Print results:
    print "I_exact =",Iexact
    print "I_hitormiss =",I
    print "Error on I =",Ierr,"%"
    
    #Replot function
    plt.plot(x,f(x),lw=3)
    plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
    plt.xlabel(r'$x$',fontsize=20)
    plt.xlim(0.,15.)

    #Plotting all of them is not recommended
    #plt.plot(samples_x,samples_y,'g.')

    #Just show the first 500 points or it gets hard to seee
    plt.plot(samples_x[:500],samples_y[:500],'g.')

    #Use the mask to show the ones below the function in red
    plt.plot(samples_x[newmask[:500]],samples_y[newmask[:500]],'r.')

    #plot box
    plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
    #vertical left
    plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
    #vertical right
    plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
    #horizontal top

    plt.title(r'$MC \ Integration$', fontsize=18)
    plt.show()
    
v = interact(hitOrMiss,N=(1000,100000,100),xmin=(1.,15.,0.001),xmax=(0.,15.,0.001))
```

### Sample Mean Method

The mean-value theorem of calculus states that the definite integral of $(a,b)$ is determined by the average value of the integrand $f(x)$ in the range $x\in(a,b)$. We sample the $x_i$ at random instead of at regular interrvals, and compute values of $f(x_i)$. For the one-dimensional integral $\int f(x) dx$, the estimate of the integral in the sample mean method is given by $I_N = (b-a)\left<f\right>$ with $\left<f\right>$ the usual definition of the mean: \begin{equation}\left<f\right>= \frac{1}{N} \sum_{i=1}^N f(x_i)\end{equation}
We can estimate the uncertainty using \begin{equation} unc_N = (b-a) \sqrt{\frac{\left<f^2\right> - \left<f\right>^2}{N}} \end{equation}

Let's integrate $f(x) = x^{1/2} + \sin(x)$ and its uncertainty using the sample mean method for $N = 10000$.

```python
def sampleMean(N=10000,xmin=1.7,xmax=12.4):
    
    #set up func
    f = lambda x: x**(0.5) + np.sin(x)
    x = np.arange(0.,15.,0.001)
    Iana = lambda x: (2./3.)*x**(3./2.) - np.cos(x)
    Iexact = Iana(xmax) - Iana(xmin)
    
    ########################################
    #Integration Part
    samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
    
    approx = f(samples_x)
    I = approx.mean()*(xmax-xmin)
    
    #Numerical Uncertainty
    Iunc = (xmax-xmin)*np.sqrt(((approx**2).mean() - (approx.mean())**2))
    
    #Comparison with exact
    Ierr = 100.*np.abs(Iexact-I) / Iexact
    #########################################
    
    #find max value of f within boundary to set ymax
    ymax = approx.max()
    ymin = 0.
    
    #Print results
    print "I_exact =", Iexact
    print "I_samplemean =", I, " +/- ",Iunc
    print "Error on I =", Ierr, "%"
    
    #Plot function
    plt.plot(x,f(x),lw=3)
    plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
    plt.xlabel(r'$x$',fontsize=20)
    plt.xlim(0.,15.)
    
    #plot box
    plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--')
    #vertical left
    plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--')
    #vertical right
    plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
    #horizontal top
    
    #Plot average value of function from samples to compare
    plt.plot((xmin,xmax),(approx.mean(),approx.mean()),'r-',label=
             "Average value of function at sample points")
    plt.legend(loc="upper left")
    
    plt.title(r'$MC \ Integration$', fontsize=18)
    plt.show()
    
v2 = interact(sampleMean,N=(1000,100000,100),xmin=(0.,15.,0.001),
              xmax=(0.,15.,0.001))
```

## Multi-Dimensional Integrals

In 3-d, we can solve
\begin{equation}
\int_0^{0/10} dx \int_0^1 dy \int_0^{11/10} dz (4 - x^2 - y^2 - z^2) = \frac{14817}{5000} = 2.9634
\end{equation}

```python
fxyz = lambda x,y,z: 4. - x**2 - y**2 - z**2
xa = 0.; xb = 9./10.
ya = 0.; yb = 1.
za = 0.; zb = 11./10.

i = (10,100,1000,10000,100000,1000000)
for N in i:
    x = (xb-xa)*np.random.random_sample(N)
    y = (yb-ya)*np.random.random_sample(N)
    z = (zb-za)*np.random.random_sample(N)
    
    I = (xb-xa)*(yb-ya)*(zb-za)*fxyz(x,y,z).mean()
    err = (xb-xa)*(yb-ya)*(zb-za)*np.sqrt(((fxyz(x,y,z)**2).mean() - (fxyz(x,y,z).mean())**2)/N)
    exact = 14817./5000
    
    print "Exact =",exact,"I=",I," +/-",err,"percentErr = ",100.*np.abs(I-exact)/exact
```

Let's try one in five dimensions:

\begin{equation}\int_0^{1/10} dx_1 \int_0^{4/5} dx_2 \int_0^{9/10} dx_3 \int_0^1 dx_4 \int_0^{11/10} dx_5 (6-x_1^2-x_2^2-x_3^2-x_4^2-x_5^2) = \frac{63987}{25000}=2.55948\end{equation}

```python
func5D =  lambda x1,x2,x3,x4,x5: 6 - x1**2 - x2**2 - x3**2 - x4**2 - x5**2

L = 7./10; W = 4./5; H = 9./10; P = 1.; Q = 11./10

exact = 63987./25000

for N in (10,100,1000,10000,100000,1000000):
    x1 = L*np.random.random_sample(N)
    x2 = W*np.random.random_sample(N)
    x3 = H*np.random.random_sample(N)
    x4 = P*np.random.random_sample(N)
    x5 = Q*np.random.random_sample(N)

    I = (L*W*H*P*Q)*func5D(x1,x2,x3,x4,x5).mean()
    unc = L*W*H*P*Q*np.sqrt((func5D(x1,x2,x3,x4,x5)**2).mean() - (func5D(x1,x2,x3,x4,x5).mean())**2)
    err = 100*np.abs(I - exact)/exact
    
    print "Integral=%.5f, +/- %.6f, percent err = %.3f, N = %d"%(I,unc,err,N)
```

And finally for a function with no exact solution:
\begin{equation}
\int_0^{1/10} dx_1 \int_0^{4/5} dx_2 \int_0^{9/10} dx_3 \int_0^1 dx_4 \int_0^{11/10} dx_5 \sqrt{6-x_1^2-x_2^2-x_3^2-x_4^2-x_5^2}
\end{equation}

```python
func5D =  lambda x1,x2,x3,x4,x5: np.sqrt(6 - x1**2 - x2**2 - x3**2 - x4**2 - x5**2)

L = 7./10; W = 4./5; H = 9./10; P = 1.; Q = 11./10

for N in (10,100,1000,10000,100000,1000000):
    x1 = L*np.random.random_sample(N)
    x2 = W*np.random.random_sample(N)
    x3 = H*np.random.random_sample(N)
    x4 = P*np.random.random_sample(N)
    x5 = Q*np.random.random_sample(N)

    I = (L*W*H*P*Q)*func5D(x1,x2,x3,x4,x5).mean()
    unc = L*W*H*P*Q*np.sqrt((func5D(x1,x2,x3,x4,x5)**2).mean() - (func5D(x1,x2,x3,x4,x5).mean())**2)
    
    print "Integral=%.5f, +/- %.6f, N = %d"%(I,unc,N)
```

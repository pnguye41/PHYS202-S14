## The Simple Pendulum as an example

If you use Newton's second law to write the sum of the forces on the mass, *m*, you get $\sum \vec{F} = \vec{T} + \vec{w}$, which is a vector expression that needs to be broken up into its components. The radial acceleration is zero, so we need only consider tangential acceleration. It makes most sense to use cylindrical coordinates.

\begin{equation}
F_{radial} = T - mgcos \theta = ma_{radial} = 0
\\
F_{tangential} = -mgsin \theta = ma_{tangential}
\end{equation}

$a_{tangential}$ is related to $\theta$ through the arc-length *s* that the mass makes as it swings:

\begin{equation}
s=l\theta
\\
v_{tangential} = \frac{ds}{dt} = l \frac{d \theta}{dt}
\\
a_{tangential} = \frac{d^2s}{dt^2} = l \frac{d^2 \theta}{dt^2}
\end{equation}

which gives the following D.E. that we need to solve
\begin{equation}
\frac{d^2 \theta}{dt^2} = -\frac{g}{l}sin \theta
\end{equation}

Due to the $sin \theta$ on the right-hand-side, this has no analytic solution. So we only have two possible paths forward:

1. Approximate
2. Solve numerically

### Approximate solution

Recall \begin{equation}
sin\theta = \theta - \frac{\theta ^3}{3!} + \frac{\theta ^5}{5!} - \frac{\theta ^7}{7!} + ...
\end{equation}
In closed form:
\begin{equation}
\frac{d^2 \theta}{dt^2} = -\frac{g}{l}\theta
\end{equation}
Which has solution \begin{equation}
\theta(t)=\theta_0cos(\sqrt{g/l}t)
\end{equation}

```python
g = 9.8 #m/w^2
ell = 0.5 #m
th0 = 20.*np.pi/180 # 20 degrees, but must use radians

t = np.arange(0.,10.+0.01,0.01)
theta_app = lambda t: th0*np.cos(np.sqrt(g/ell)*t)

plt.plot(t,theta_app(t),label='Approx. solution')
plt.xlabel(r'$t (s)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

### Numerical Solution

We need to convert our single 2nd-order D.E. into two 1st-order equations. Generally, we use $\omega = \frac{d\theta}{dt}$ to represent angular velocity.
\begin{equation}
\frac{d^2\theta}{dt^2} = \frac{d\omega}{dt} = -\frac{g}{l}sin\theta
\end{equation}
So now we have two first order equations:
\begin{equation}
\left[\frac{d\theta}{dt} = \omega, \frac{d\omega}{dt} = -\frac{g}{l}sin\theta\right]
\end{equation}
The function $\theta(t)$ will be the first element `theta[0]` and the derivative $\omega$ will be `theta[1]`.

```python
def deriv(theta_current,t): #return derivatives of array theta
    g = 9.8 # m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return np.array([dtheta, domega])
```

The `odeint` solver will take an array of current values for $[\theta,\omega]$ and return values for $[d\theta/dt,d\omega/dt]$ at each time step.

```python
from scipy.integrate import odeint
```

```python
t = np.arange(0.,10.+0.01,0.01)

omega0 = 0. #starting from rest, omega = 0
theta_init = np.array([th0, omega0]) # initial values, th0 from before = 20 degrees, in radians

#pass function deriv to the solve with initial values and integration variable
theta = odeint(deriv,theta_init,t)
print theta.shape
```

```python
plt.plot(t,theta_app(t),'b-',label="Approx solution")
plt.plot(t,theta[:,0],'r--',label="Num solution")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

```python
th0 = 50.*np.pi/180 # 50 degrees

theta_init = np.array([th0,omega0]) #initial values
theta = odeint(deriv,theta_init,t)

plt.plot(t,theta_app(t),'b-',label="Approx solution")
plt.plot(t,theta[:,0],'r--',label="Num solution")
plt.xlabel(r'$t (seconds)$',fontsize=20)
plt.ylabel(r'$\theta (rad)$',fontsize=20)
plt.ylim(-1,1.5)
plt.grid()
plt.legend()
plt.show()
```

### Making physics come alive

```python
from IPython.html.widgets import interact, interactive
```

```python
def derivSP(thetaIn,t,g,ell): #return derivatives of the array theta
    theta = thetaIn[0]
    omega = thetaIn[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return array([dtheta, domega])

def simplePendulum(th0=20.,w0=0.,ell=0.5,g=9.8):
    
    t = np.arange(0.,10.+0.01,0.01)
    
    #Approx analytic solution for comparison
    approx = th0*np.pi/180*np.cos(np.sqrt(g/ell)*t)
    
    #convert deg to rad
    initVals = np.array([th0*np.pi/180,w0])
    
    #pass function deriv to solver with init values and integration variable
    theta = odeint(derivSP,initVals,t,args=(g,ell))
    #theta[:,0] are positions
    #theta[:,1] are angular velocities
    
    #Plot results
    plt.plot(t,approx,label='Approx solution')
    plt.plot(t,theta[:,0],label='Num solution')
    plt.xlabel(r'$t (sec)$',fontsize=20)
    plt.ylabel(r'$\theta (rad)$',fontsize=20)
    plt.ylim(-2.,2.)
    plt.grid()
    plt.legend()
    plt.show()
    
v = interact(simplePendulum,th0=(0.,90.),w0=(0.,10.),ell=(0.1,1.0),g=(1.0,20.0))
```

## Breaking into the black box

### Euler's Method

The most basic numerical algorithm for computing solutions to ODEs was invented by Euler.

### Free-fall as an example

\begin{equation}
a(t) = \frac{d^2y}{dt^2}
\\
a(t) = \frac{dv}{dt}, v(t) = \frac{dy}{dt}
\end{equation}

Starting at a time t0 at some known position y(t0) and with some known velocity v(t0), we can calculate the change in velocity and position in going from a time t0 to a time t1=t0+Δt
\begin{equation}
v(t_1) = v(t_0) + a(t_0)\Delta t
\\
y(t_1) = y(t_0) + v(t_0)\Delta t
\end{equation}
This is the Euler Method. It is based on a tangent line approx. to the function in question.

Let's denote v(t1) as v1 and so on. Applying the method successively on subsequent time steps up to n we can write ti+1=ti+Δt and ti=t0+iΔt, so that

\begin{equation}
v_{i+1}=v_i + a_i\Delta t
\\
y_{i+1}=y_i + v_i\Delta t
\end{equation}

```python
def EulerFreefall():
    tmax = 10 #max time
    dt = 0.1 #time step
    a = -9.81 #g in m/s^2
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    #set init conditions, redundant in this example
    t[0] = 0.
    v[0] = 0.
    
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i]
        
    return t,v,y

te,ve,ye = EulerFreefall()
```

```python
plt.plot(te,ye,"r-",label="y(t)")
plt.plot(te,ve,"b-",label="v(t)")
plt.xlabel('t (s)',fontsize=20)
plt.ylabel('y(t),v(t)',fontsize=20)
plt.grid()
plt.legend(loc='best',fontsize=20)
plt.show()
```

### Exploring the parameter space

Consider ball starting from y=0 with init velocity upwards of 46 m/s.
Let's set this up with an `interact` object so we can explore the dynamics for different init conditions without writing lots of code. Let's compare the numerical result to an analytical one.

```python
def freefall(tmax=10.,dt=0.1,t0=0.,v0=46.,y0=0.):
    a = -9.81 #g in m/s^2
    
    #Set up arrays based on time parameters
    t = np.zeros(int(tmax/dt))
    v = np.zeros(int(tmax/dt))
    y = np.zeros(int(tmax/dt))
    
    #set up init conditions based on parameters
    t[0] = t0
    v[0] = v0
    y[0] = y0
    
    #Apply Euler's method to solve
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i]
        
    #Compare to analytical result for ymax
    tmax = -v0/a
    ymax = y0 + v0*tmax + 0.5*a*tmax**2
    percentDiff = 100.*np.abs(y.max()-ymax)/ymax
    print 'ymax percenDiff = %.2f'%percentDiff
    #plot it
    plt.plot(t,y,"r-",label="y(t)")
    plt.plot(t,v,"b-",label="v(t)")
    plt.xlabel("time (s)",fontsize=20)
    plt.ylabel("y(t),v(t)",fontsize=20)
    plt.xlim(0,20.)
    plt.ylim(-100.,600.)
    plt.grid()
    plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
    plt.show()
```

```python
v4 = interact(freefall,tmax=(1.,20.),dt=(0.01,1.,0.01),t0=(0.,20.),v0=(10.,100.),y0=(0.,10.0))
```

### Euler-Cromer Method

Varying timespteps let's us improve the accuracy of our result, but to do an even better job, we could try modifying the method for calculating the new positions and velocities. Cromer added the idea of calculating the new position using the just calculated new velocity (at the end of the interval) to Euler's method, creating the Euler-Cromer Method.
\begin{equation}
v_{i+1}=v_i + a_i\Delta t
\\
y_{i+1}=y_i + v_{i+1}\Delta t
\end{equation}

```python
def EulerCromerFreefall():
    tmax = 10 #max time
    dt = 0.1 #time step
    a = -9.81 #g in m/s^2
    nsteps = int(tmax/dt)
    t = np.zeros(nsteps)
    v = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    # set init conditions
    t[0] = 0.
    v[0] = 0.
    
    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v[i+1] = v[i] + dt * a
        y[i+1] = y[i] + dt * v[i+1]
        
    return t,v,y
```

```python
tec,vec,yec = EulerCromerFreefall()
```

```python
plt.plot(tec,yec,"r-",label="y(t)")
plt.plot(tec,vec,"b-",label="v(t)")
plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),v(t)",fontsize=20)
plt.grid()
plt.title("Euler-Cromer")
plt.legend(loc="best", fontsize=20)
plt.show()
```

### Midpoint Method

Even more accurate than previous two b/c it uses slope at midpoint of time interval to estimate forward progress of velocity.
\begin{equation}
v_{i+1} = v_i + a_i \Delta t
\\
y_{i+1} = y_i + \frac{1}{2}(v_i + v_{i+1}) \Delta t
\end{equation}

### Runge-Kutta Family of Methods

These are a series of numerical methods for solving DEs and systems of DEs of which Euler and the Midpoint methods are a subset. Euler is an order 1 method, Midpoint order 2, and RK4, the most commonly used, order 4.

RK4 takes the form

\begin{equation}
y_{i+1} = y{i} + \frac{\Delta t}{6}(k_i + 2k_2 + 2k_3 + k_4)
\end{equation}

where 

\begin{equation}
k_1 = f(y_i,t_i)
k_2 = f(y_i + \frac{\Delta t}{2}k_2, t_i + \frac{1}{2}\Delta t)
k_3 = f(y_i + \frac{\Delta t}{2}k_3, t_i + \frac{1}{2}\Delta t)
k_4 = f(y_i + \Delta tk_3, t_i + \Delta t)
\end{equation}

fro an initial value problem specified by $\frac{dy}{dt} =f(y,t)$, with y(t0) = y0.

$y_{i+1}$ is the RK4 approximation of $y(t_{i+1})$, and the next value ($y_{i+1}$) is determined by the present value ($y_i$) plus the weighted average of four increments, where each increment is the product of the size of the interval, $\Delta t$, and an estimated slope specified by function $f$ on the right-hand side of the DE.

### Coupled differential equations

A predator-prey model describes the interactions of two species in an ecosystem. For this example, let’s consider lions and zebras. If we consider a group of lions and zebras that don’t interact with any species but each other, we can write a system of equations to govern their populations as follows:

\begin{equation}
\frac{dZ}{dt} = \alpha Z - \beta ZL,
\\
\frac{dL}{dt} = \epsilon \beta ZL - \gamma L,
\end{equation}

where L and Z are populations of lions and zebras. Equations represent rate of change of populations given rates of birth and death of each species.

Death rates are controlled by rate of lions consuming zebras and starving to death from lack of zebras. Init conditions are starting populations: $L(0) = L_0, Z(0) = Z_0.$
The parameters are
* $\alpha$: growth rate of zebras [1/years]
* $\gamma$: death rate of lions [1/years]
* $\beta$: death rate of zebras per encounter with lion [1/lions/years]
* $\epsilon$: efficiency of producing lions from consumed zebras [linos/zebras]

With L = 0:
\begin{equation}
\frac{dZ}{dt} = \alpha Z
\\
Z(t)Z_0 e^{\alpha t}
\end{equation}

With Z = 0:
\begin{equation}
\frac{dL}{dt} = -\gamma L
\\
L(t) = L_0 e^{-\gamma t}
\end{equation}

```python
#we want to solve system dy/dt = f(y,t)
def PPderiv(y,t):
    
    alpha = 0.7
    gamma = 0.5
    beta = 0.007
    eps = 0.1
    
    Zi = y[0]
    Li = y[1]
    
    dZdt = alpha*Zi - beta*Zi*Li
    dLdt = eps*beta*Zi*Li - gamma*Li
    
    return np.array([ dZdt, dLdt ])
```

```python
#50 year span
t = np.arange(0.,50.+0.1,0.1)
#initial conditions
Z0 = 200
L0 = 50

#pass func deriv to solver with init values and integ variable
result = odeint(PPderiv, np.array([Z0,L0]),t)

Z = result[:,0]
L = result[:,1]

plt.plot(t,Z,label="Zebra population")
plt.plot(t,L,label="Lion population")
plt.xlabel("time (years)")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
res1 = odeint(PPderiv,np.array([100,50]),t)
res2 = odeint(PPderiv,np.array([300,50]),t)
plt.plot(res1[:,0],res1[:,1],label="Z$_0$ = 100, L$_0$ = 50")
plt.plot(Z,L,label="Z$_0$ = 200, L$_0$ = 50")
plt.plot(res2[:,0],res2[:,1],label="Z$_0$ = 300, L$_0$ = 50")
plt.xlabel("Zebras",fontsize=20)
plt.ylabel("Lions",fontsize=20)
plt.legend()
plt.show()
```

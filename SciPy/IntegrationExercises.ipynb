**(a)** Write two functions `trapz(func,x)` and `simps(func,x)` to compute the integral of the function func over the variable `x` using the trapezoidal rule and Simpson's rule to a file called `Integrators.py`. Do not use the scipy.integrate built-in functions. Include docstrings with each function that describe what they do. Then import the module and use the functions to answer the following questions.

```python
%%file Integrators.py

import numpy as np

def trapz(func,x,lower_bound,upper_bound,slices):
    """Approximates integral of function 'func' of variable 'x' between lower and upper bounds\
    by summing the areas of N trapezoidal slices of equal widths.
    ***Lower and Upper bound arguments must be of type float, not int.***
    """
    N = slices
    a = lower_bound
    b = upper_bound
    h = (b-a)/N

    k = np.arange(1,N)
    I = h*(0.5*func(a) + 0.5*func(b) + func(a+k*h).sum())

    print "Trapezoidal Rule Integral =", I
    return I

def simps(func,x,lower_bound,upper_bound,slices):
    """Approximates integral of function 'func' of variable 'x' between lower and upper bounds\
    by fitting parabolas to the curve and summing the areas of N equal width slices.
    ***Lower and Upper bound arguments must be of type float, not int.***
    """
    N = slices
    a = lower_bound
    b = upper_bound
    h = (b-a)/N

    k1 = np.arange(1,N/2+1)
    k2 = np.arange(1,N/2)
    I = (1./3.)*h*(func(a)+func(b) + 4.*func(a+(2*k1-1)*h).sum() + 2.*func(a+2*k2*h).sum())

    print "Sumpson's rule Integral =",I
    return I
```

## Question 2

Consider the integral

\begin{equation}
E(x)=\int_0^xe^{−t^2}dt
\end{equation}

This is the error function, commonly seen in probability and statistics. There is no known way to perform this particular integral analytically, although the integrand can be expanded in a Taylor series and terms computed to arbitrary order. For most applications, numerical approaches are the only way forward.

**(a)** Use `scipy.integrate.cumtrapz` to calculate $E(x)$ for values of $x$ from 0 to 3 in steps of 0.1. Print the result.

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.0,3.0,0.1)
t = x
y = np.e**((-1)*(t**2))
E = integrate.cumtrapz(y, x, initial=0)
print E
```

**(b)** Plot the integrand as a function of t and E(x) as a function of x (both ranging from 0 to 3) on the same graph.

```python
plt.figure(figsize=(11,7))
plt.plot(t,y,lw=3,label='Integrand ($e^{-t^2}$)')
plt.plot(x,E,lw=3,label='Error Function $E(x)$')
plt.title('Error function and Integrand',fontsize=20)
plt.xlabel('x and t',fontsize=16)
plt.ylabel('$E(x)$ and $e^{-t^2}$',fontsize=16)
plt.legend(loc='right')
plt.show()
```

Consider a particle in a one-dimensional box of width L. The probability of finding the particle between a and b is given by

\begin{equation}
P(a,b)=\int_b^a|\psi(x)|^2dx
\end{equation}

where

\begin{equation}
\psi(x)=\sqrt{\frac{2}{L}}sin\left(\frac{n \pi x}{L}\right)
\end{equation}

is the wavefunction.

**(a)** What is the probability of finding the particle between $L/3$ and $L/2$ for the ground state ($n = 1$) and for the first excited state ($n = 2$)? Let $L = 1$. Perform the integral using both `scipy.integrate.trapz` and `scipy.integrate.quad` with an accuracy of 6 sig figs. How many slices did you need to use for `scipy.integrate.trapz`?

```python
def integration_test(N=1000):
    x = np.linspace(1./3.,1./2.,N)
    psi1_trapz = (np.sqrt(2)*np.sin(1.*np.pi * x))**2
    result1 = integrate.trapz(psi1_trapz,dx=1/(N*6.))
    print "n = 1, using trapz:\n",result1

    psi2_trapz = (np.sqrt(2)*np.sin(2.*np.pi * x))**2
    result2 = integrate.trapz(psi2_trapz,dx=1/(N*6.))
    print "\nn = 2, using trapz:\n",result2

    psi1_quad = lambda x : (np.sqrt(2)*np.sin(1.*np.pi * x))**2
    result3,error = integrate.quad(psi1_quad,(1/3.),(1/2.))
    print "\nn = 1, using quad:\n",result3

    psi2_quad = lambda x : (np.sqrt(2)*np.sin(2.*np.pi * x))**2
    result4,error = integrate.quad(psi2_quad,(1/3.),(1/2.))
    print "\nn = 2, using quad:\n",result4

from IPython.html.widgets import interact, interactive
widget = interact(integration_test,N=(100000,1000000,100))
```

## Question 4

Integrals of mass density lead to three interesting quantities:

\begin{equation}
M = \int \rho dV
\\
\vec{r}_{cm} = (x_{cm},y_{cm},z_{cm}) = \frac{1}{M}\int \vec{r} \rho dV
\\
I_{cm} = \int (r - r_{cm})^2 \rho dV
\end{equation}

where $M$ is the total mass (a scalar), $\vec{r}_{cm}$ is the center of mass position (a vector of 3 components), $I_{cm}$ are the moments of inertia about the center of mass (diagonals of a 3x3 matrix), and the mass density, $\rho$, may be a function of the spatial variables.

Consider a rectangular box: length (in x) = 0.2m, width (in y) = 0.2m, and height (in z) = 1.0m centered on the origin, (0,0,0) and with a mass density, 

$\rho (x,y,z) = (100 $ kg/m^4 $)(y + 0.1) + (100$ kg/m^5 $)z^2$ 

for x, y, and z in meters.

**(a)** Use numerical integration to find $M$, $\vec{r}_{cm}$, and $I_{cm}$. Note that you need M to compute r_cm and r_cm to compute I_cm. The moments of inertia $I_{cm}$ are $I_{xx}$,$I_{yy}$, and $I_{zz}$, where e.g. $I_{xx}=\int((y-y_{cm})^2 + (z-z_{cm})^2)\rho dV$, etc.

```python
import numpy as np
from scipy import integrate

integrand_density = lambda x,y,z : 100.*y + 10. + (100.*(z**2))

xmin = 0.
xmax = 0.2

ymin = lambda x : 0.
ymax = lambda x : 0.2

zmin = lambda x,y : 0.
zmax = lambda x,y : 1.

M,Merr = integrate.tplquad(integrand_density, xmin, xmax, ymin, ymax, zmin, zmax)
print "M =",M

integrand_xcm = lambda x,y,z : x * integrand_density(x,y,z)
xcm,xcmerr = integrate.tplquad(integrand_xcm, xmin, xmax, ymin, ymax, zmin, zmax)
print "\n\nx_cm =",xcm

integrand_ycm = lambda x,y,z : y * integrand_density(x,y,z)
ycm,ycmerr = integrate.tplquad(integrand_ycm, xmin, xmax, ymin, ymax, zmin, zmax)
print "y_cm =",ycm

integrand_zcm = lambda x,y,z : z * integrand_density(x,y,z)
zcm,zcmerr = integrate.tplquad(integrand_zcm, xmin, xmax, ymin, ymax, zmin, zmax)
print "z_cm =",zcm

rcm = [xcm,ycm,zcm]
print "\nr_cm =",rcm

integrand_Ixx = lambda x,y,z : ((y - ycm)**2 + (z - zcm)**2) * integrand_density(x,y,z)
Ixx,Ixxerr = integrate.tplquad(integrand_Ixx, xmin, xmax, ymin, ymax, zmin, zmax)
print "\n\nI_xx =",Ixx

integrand_Iyy = lambda x,y,z : ((x - xcm)**2 + (z - zcm)**2) * integrand_density(x,y,z)
Iyy,Iyyerr = integrate.tplquad(integrand_Iyy, xmin, xmax, ymin, ymax, zmin, zmax)
print "I_yy =",Iyy

integrand_Izz = lambda x,y,z : ((x - xcm)**2 + (y - ycm)**2) * integrand_density(x,y,z)
Izz,Izzerr = integrate.tplquad(integrand_Izz, xmin, xmax, ymin, ymax, zmin, zmax)
print "I_zz =",Izz
```
